```python
import pandas as pd
import random

w = []
y = []
for i in range(1,10000):
    v = random.random()
    w.append(v)
    
w.sort(reverse=True)

for i in range(1,len(w)):
    if ( random.random() > w[i] ):
        y.append(w[i])
    
print(len(y))
s = pd.Series(y)
s.hist(bins=10)
```
